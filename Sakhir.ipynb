```python
from matplotlib import pyplot as plt

import seaborn as sns
import pandas as pd

import F1
```

```python
season = F1.Season(2020)
season.racename
```

```python
round = 16
```

```python
season.create_race(round)
season.races[round].name
```

```python
season.races[round].get_timing_data()
```

```python
df = season.races[round].df_timing
```

```python
sns.set(rc={'figure.figsize':(32,20)})
sns.boxplot(x='Driver', y='LapTime', data=df)
sns.stripplot(x='Driver', y='LapTime', data=df, color=".3")
plt.title(season.races[round].name)
plt.ylabel('LapTime [min]')
```

```python
sns.boxplot(x='Constructor',y='LapTime', data=df)
plt.ylabel('LapTime [min]')
```

```python
sns.violinplot(x='Driver', y='LapTime',
 data=df.loc[df['Constructor'] == 'mercedes'])
plt.ylabel('LapTime [min]')
```

```python
season.races[round].df_timing.groupby(['Driver'])['LapTime'].mean()
```

```python
sns.lineplot(x='Lap', y='LapTime', data=df.loc[df['Constructor'] == 'mercedes'], hue='Driver')
plt.ylabel('LapTime [min]')
```
